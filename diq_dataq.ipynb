INITIAL DATA QUALITY ASSESSMENT

```python
import pandas as pd
import seaborn
```

```python
# Load the dataset
db = pd.read_csv('Comune-di-Milano-Esercizi-di-vicinato-in-sede-fissa.csv',sep=';',encoding='unicode_escape')
```

```python
# Display the shape of the dataset
db.shape
```

```python
# Display the column names of the dataset
db.columns
```

```python
# Display the data types of each column
db.dtypes
```

```python
# Identify duplicate rows in the dataset
DUPLICATES = db.duplicated()
DUPLICATES
```

```python
# Check for duplicate rows
print(DUPLICATES.any())
```

```python
# Display duplicate rows
db[DUPLICATES]
```

```python
# Show the total number of rows in the dataset for each column
db.count()
```

```python
# Show the number of missing values in each column
db.isnull().sum()
```

```python
# Show the total number of non-null values in the dataset
NOT_NULL = db.count().sum()
db.count().sum()
```

```python
# Show the total number of null values in the dataset
NULL = db.isnull().sum().sum()
db.isnull().sum().sum()
```

```python
# Calculate and display the completeness of the dataset
TOT = NULL + NOT_NULL
COMPLETENESS = NOT_NULL / TOT

COMPLETENESS = "{0: .1f}%".format(COMPLETENESS*100)
print(COMPLETENESS)
```

```python
# Display summary statistics of the dataset
db.describe()
```

```python
# Calculate and display the Pearson correlation matrix for numerical columns
NUM = list(db.select_dtypes(include=['int64','float64']).columns)
db[NUM].corr(method='pearson')
```

```python
# Calculate and display the Spearman correlation matrix for numerical columns
corr = db[NUM].corr(method ='spearman')
# Display the correlation matrix as a heatmap
seaborn.heatmap(corr)
```
